```python
#Reverse a linked list and Palindrome

class Node:
    
    def __init__(self,data):
        self.data = data 
        self.next = None
    
def insert(head,data):
    node = Node(data)
    node.next = head
    return node

def reverse(head):
    current = head
    prev = None
    while(current is not None):
        temp = current.next
        current.next = prev
        prev = current
        current = temp
        
    return prev
        
            
def traverse(head):
    current = head 
    while current:
        print(current.data,end="->")
        current = current.next
    print("None")
            
def copy_list(head):
    newhead = None
    if head is not None:
        newhead = Node(head.data)
        newhead.next = copy_list(head.next)
    return newhead

def palindrome(head,revhead):   
    flag =0 
    while(head is not None and revhead is not None):
        if(head.data != revhead.data):
            print("Not Palindrome")
            flag = 1
            break
        head = head.next 
        revhead = revhead.next
        
    if(flag == 0):
        print("Palindrome")
        

head = None

head = insert(head,1)
head = insert(head,2)
head = insert(head,2)
head = insert(head,1)

head_copy = copy_list(head)
traverse(head)
head2 = reverse(head)
traverse(head2)
palindrome(head_copy,head2)
```

```python
#Reverse Doubly Linked List using recursion

class Node:
    def __init__(self,data):
        self.data = data
        self.prev = None
        self.next = None
        
def begin_insert(head,data):
    
    new = Node(data)
    new.prev = None
    new.next = head
    if head is not None:
        head.prev = new        
    return new

def traverse_front(head):
    current = head 
    while current:
        print(current.data,end="<->")
        current = current.next
    print("None")

def rev(head):
    if head is None:
        return None

    temp = head.next
    head.next = head.prev
    head.prev = temp
    
    if not head.prev:
        return head
    return rev(head.prev)
    

    
head = None

head = Node(5)
head.next = Node(4)
head.next.prev = head
head.next.next = Node(3)
head.next.next.prev = head.next
head.next.next.next = None

traverse_front(head)
head2 = rev(head)
traverse_front(head2)
```

```python
#Sort Doubly Linked List using quick sort

#Incomplete Program
class Node:
    def __init__(self,data):
        self.data = data
        self.prev = None
        self.next = None

def insert(head,data):
    
    new = Node(data)
    new.prev = None
    new.next = head
    if head is not None:
        head.prev = new        
    return new

def traverse_front(head):
    current = head 
    while current:
        print(current.data,end="<->")
        current = current.next
    print("None")

def convert_to_list(head):
    current = head
    list1 = []
    while current:
        list1.append(current.data)   
        current = current.next
    return list1

def quick_sort(list1,low,high):
    if low<high:
        
        p = part(list1,low,high)
    
        quick_sort(list1,low,p-1)
        quick_sort(list1,p+1,high)
    
def part(list1, low, high):
    
    pivot = list1[high]
    i = low - 1
    
    for j in range(low,high):
        if list1[j]<pivot:
            i +=1
            swap(list1,i+1,high)
    
    swap(list1, i + 1, high)
    return i + 1


    
def swap(list1, i, j):
    list1[i], list1[j] = list1[j], list1[i]


head = None
head = insert(head,8)
head = insert(head,3)
head = insert(head,7)
head = insert(head,1)

list1 = convert_to_list(head)
quick_sort(list1,0,len(list1)-1)
print(list1)
traverse_front(head)
```
