```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car = sns.load_dataset("car_crashes")
car.head()
```

```python
sns.lineplot(x="alcohol", y="speeding",data = car,size = 10)
plt.show()
```

```python
diamond = sns.load_dataset("diamonds")
diamond = diamond.head(20)
diamond
```

```python
sns.lineplot(x="carat", y="price",data = diamond,size=10,hue="cut", style = "cut",palette="husl",markers=["o",">"],dashes = False)
plt.show()
```

```python
sns.color_palette("husl")
```

```python
sns.lineplot(x="carat", y="depth",data = diamond,size=10,hue="clarity", style = "clarity",palette="Accent",markers=["o",">"],dashes = False,legend = False)
plt.grid()
plt.title("Diamonds")
plt.show()
```

```python
sns.histplot(diamond["color"],kde = True, color = 'y')
plt.show()
```
